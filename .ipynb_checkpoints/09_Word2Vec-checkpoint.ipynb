## Word2Vec - CBOW and Skip-Gram

- Word2Vec is a method to learn word embeddings, where words are represented as dense vectors in a continuous space. It captures semantic similarities between words based on their usage in a large corpus.

    - CBOW (Continuous Bag of Words): Predicts a target word based on its surrounding context words. It aggregates information from neighboring words to infer the central word. For example, given the context "the cat is on the," CBOW tries to predict "mat."

    - Skip-Gram: Does the opposite of CBOW. It predicts surrounding context words given a target word. For instance, given "mat," it tries to predict context words like "the," "cat," and "on." Skip-Gram works well with smaller datasets and captures rare word representations better.

- Both methods produce word embeddings that capture semantic relationships, and Word2Vec is widely used in NLP tasks like text classification, sentiment analysis, and machine translation.
